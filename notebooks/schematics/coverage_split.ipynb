```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt
from crl.utils.graphing import despine
```

# Split Conformal Coverage Beta variation

Recreation of the Beta distribution plot in [Tibshirani's tutorial](https://www.stat.berkeley.edu/~ryantibs/statlearn-s23/lectures/conformal.pdf)



```python
alpha = 0.1

m_vals = [50, 100, 500]

for m in m_vals:
    t = np.ceil((1 - alpha) * (m + 1))

    dist = beta(t, m + 1 - t)
    xs = np.linspace(0.8, 1, 200)
    ys = dist.pdf(xs)

    plt.plot(xs, ys, label=m)

plt.axvline(1-alpha, linestyle='--', linewidth=1, alpha=0.5, c='k')
plt.yticks([])
plt.xticks([0.8, 0.85, 0.9, 0.95, 1.0])
plt.legend(frameon=False, title='m')
plt.ylim(0, None)
plt.xlabel('Coverage')
plt.ylabel('Probability')
ax = plt.gca()
despine(ax)

plt.savefig('../../results/figures/schematics/split_conformal_coverage.pdf')
plt.show()
```
